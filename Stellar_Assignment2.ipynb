```python
#2b)
import matplotlib.pyplot as plt
import numpy as np

#units of joules and Kelvin
En=(-2.18e-18, -5.4474e-19, -2.41929e-19)
k=1.380649e-23
T_values=np.arange(2000,22000)
T=1
n_values=(1,2,3)
E1=-2.18e-18
Z1H=2
Z1=Z1H

Z2H=Z1+8*np.exp(-10.2/(k*T))
Z2=Z2H

Z3H=Z2+18*np.exp(-12.09/(k*T))
Z3=Z3H
```

```python
#units are joules and kg

#Ionised H:


ne=96e19          #my student number (21365396) ends in 96 => number density of free electrons=96*10^-19
me=9.1093837e-31  #mass of electron
h=6.62607015e-34  #planck constant
XH=2.18e-18       #ionisation energy of hydrogen
X=XH

def saha2(T):
  return ((2*k*T*Z2)/(ne*k*T*Z1))*(((2*np.pi*me*k*T)/(h**2))**1.5)*np.exp(-X/(k*T))

plt.plot(T_values,saha2(T_values)/(saha2(T_values)+1))
plt.title('Percentage of Ionised H at a Given Temperature T')
plt.ylabel('Percentage of Ionised H')
plt.xlabel('Temperature (K)')
plt.axhline(y = 0.9, color='r',linestyle='dashed', label='90% Ionisation') # 2c) to estimate where 90% is ionised
plt.axvline(x = 11300, color='maroon',linestyle='dashed')
plt.legend()
plt.grid('True')
plt.show()
```
