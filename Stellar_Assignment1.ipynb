```python
import matplotlib.pyplot as plt
import numpy as np
from astropy import units as u
from astropy import constants as c

b = 2.89777e-3
temp=8396

lambdpeak=(b/temp)/(1e-10)

print(f"according to Wien's law, peak wavelength for my star is: {lambdpeak}Å")
```

```python
lambd=np.arange(200.,20000.,1.)

T1=8396
Kb=1.380649e-23
c=2.998e8
h=6.62607015e-34
lambd1=200

FPW1=[] ##flux per wavelength of my star

for lambd1 in lambd:
  lambd2= lambd1*(1e-10)
  m=(2*np.pi*h*(c**2))/(lambd2**5)
  f=1/(np.exp((h*c)/(lambd2*Kb*T1))-1)
  FPW1.append(m*f)
print(m*f)

plt.plot(lambd,FPW1,label='"My Star"')

T2=3100
Kb=1.380649e-23
c=2.998e8
h=6.62607015e-34
lambd1=200
FPW2=[]

for lambd1 in lambd:
  lambd2= lambd1*(1e-10)
  m=(2*np.pi*h*(c**2))/(lambd2**5)
  f=1/(np.exp((h*c)/(lambd2*Kb*T2))-1)
  FPW2.append(m*f)

plt.axvline(x=lambdpeak, color='lime', linestyle='dotted',label='Peak Wavelength of "My Star"(from q2)')


plt.plot(lambd,FPW2,label='Proxima Centauri')
plt.legend()
plt.xlabel('Wavelength(Å)')
plt.ylabel('Flux(ergs s−1 cm−2 Å−1)')
plt.title('Wavelength V Flux for "my star" and Proxima Centauri')
plt.show()
```
